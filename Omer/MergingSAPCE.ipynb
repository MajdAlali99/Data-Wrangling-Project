```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
```

```python
#read the path
file_path = "../SAPCE"
file_list = os.listdir(file_path)
# locate .csv
csv_files = glob.glob('../SAPCE/*.{}'.format('csv'))
csv_files = [i[9:] for i in csv_files if 'ALL_AREAS' in i]
# merge
  
df_concat = pd.concat([pd.read_csv(f'../SAPCE/{f}') for f in csv_files ], ignore_index=True)
df_concat = df_concat.iloc[:-4].fillna(0)
```

```python
df_concat = df_concat[df_concat.TableName != 'SASUMMARY']
df_concat = df_concat[df_concat.TableName != 0]
df_concat['Region'] = df_concat['Region'].replace(' ','0')
cols = list(range(1998,2022))
cols = [str(i) for i in cols]
df_concat[cols] = df_concat[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
df_concat.drop('LineCode', axis=1).groupby('Region').sum().T.sort_index(axis=0).plot(figsize=(8,5));
```
