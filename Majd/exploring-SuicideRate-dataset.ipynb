```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
suicide = pd.read_csv('../datasets/Suicide-rate.csv')
```

```python
suicide
```

```python
suicide_rates = suicide.groupby(['YEAR', 'STATE']).sum().unstack(0)#.plot()
# plt.legend(loc=(1.04, 0));
suicide_rates
```

```python
suicide_rates.plot(kind='bar', stacked=True)
plt.xlabel('STATE')
plt.ylabel('Suicide Rates')
plt.title('Suicide Rates by State and Year')
plt.show()
```

```python
suicide_rates_transposed = suicide_rates.transpose()
# convert the year row to column
suicide_rates_transposed = suicide_rates_transposed.reset_index()
# rename the column
suicide_rates_transposed = suicide_rates_transposed.rename(columns={'index':'YEAR'})

# filter data to show only recent 3 years
suicide_rates_transposed = suicide_rates_transposed.loc[suicide_rates_transposed['YEAR'].isin(['2018','2019','2020'])]
suicide_rates_transposed

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
# plot the chart
suicide_rates_transposed.plot(x='YEAR', kind='line', stacked=True)
plt.xlabel('YEAR')
plt.ylabel('Suicide Rates')
plt.title('Suicide Rates by State and Year')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
# plot the chart
for state in suicide_rates_transposed.columns:
    if state != 'YEAR':
        plt.plot(suicide_rates_transposed['YEAR'], suicide_rates_transposed[state])
plt.xlabel('YEAR')
plt.ylabel('Suicide Rates')
plt.title('Suicide Rates by State and Year')
plt.legend(suicide_rates_transposed.columns[1:])
plt.show()
```
